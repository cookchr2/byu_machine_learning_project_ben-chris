## Data

FamilySearch shared the list of everyone who was attached to the 1900 and 1910 cenuses with us. We used these are our true matches for our training set.

We still needed a set of false matches to train our algorithm against. We took these to the censuses and binned by age, birthplace, sex, race and first letter of name and surname. We then constructed our false matches by taking the 1900 census of a true match and comparing it to everything in that bin in 1910, and vice versa. Then to add robustness, we created 3 million false matches by comparing randomly selected records. The result was 3.5 million true matches and 27 million false matches.

## Approach

We treated it like a classification problem. Our goal was to get the machine learning algorithm to determine if a pair of census records was a match or not, given the features we gave it.

We used

Ordinary Least Squares (OLS) <br>
Support Vector Machine (SVM) <br>
Random Forest (RF) <br>
Fully Connected Neural Net <br>

Note: We did not use Gradient Boosted trees, though we could have. We didn't think it would add a lot of power since our Random Forest already had overfitting issues.

## Features

We ended up using 44 features to describe our record matches. We will not list all of them here, but you can see below which ones ended up being the most important for the Random Forest

```python
image = imread("importances.png")
fig = plt.figure(figsize=(14,12))
plt.imshow(image)
plt.title("Feature Importance", {'fontsize': 22})
plt.axis("off")
plt.show()
```

### Key:

same_county: a binary variable that equals one if the records are in the same county <br>
lname_jaro: jarowinkler (string comparison) score between surnames in each record <br>
avg_jw_dim1: (jarowinkler between first word in given names + jarowinkler between first word in surnames) / 2 <br>
poi_surn_freq1910: the number of times the 1910 surname appears in the 1910 census <br>
lname_soundex: a binary variable that equals one if the surnames have the same soundex

We won't bore you with the entire list, so feel free to ask how other varaibles are defined <br>

We then did further finetuning to optimize the number of predicted matches yelided versus accuracy of predicted matches. The models above trained on an equal number of true and false matches. We discovered this is actually not what we want as the predicted matches tend to have low accuracy (between 60 and 70 percent). So, through more finetuning, we found that a 5-10:1 ratio between false and true matches in our training set gave us the best results. It was here that random forest ended up slightly outperforming the neural net. You can see how our models performed with that new ratio below.

```python
image = imread("accuracyvsnumfeatures.png")
fig = plt.figure(figsize=(18,16))
plt.imshow(image)
plt.title("Approx. Trade-off Between Accuracy and Number of Matches", {'fontsize': 24})
plt.axis("off")
plt.show()
```

So how did we do in comparison to others? Well, as you would suspect, with more training data and more possible features, we were able to significantly outperform Feigenbaum(2016). Note that he was unable to get past 90% accuracy on his testing set, while we got well over that. Also it looks like he overfit quite a bit on his random forest.

```python
image = imread("alg_compfieg.png")
fig = plt.figure(figsize=(14,12))
plt.imshow(image)
plt.title("Feigenbaum's Model Accuracy", {'fontsize': 22})
plt.axis("off")
plt.show()
```

After all the finetuning was over, we ran the random forest on our entire dataset. It ended up taking 8.5 days to run. But the end result was definitely worth the wait. Though it got roughly the same accuracy as our other models, it did significantly better at predicting out of sample.

## Predictions

Our method for finding possible matches was similiar to how we found false matches. This time though, we compared everyone in the 1900 census to everyone in the 1910 census who was in the same bin, after dropping records we knew were already matched. We then extracted the features for each match and gave them to the Random forest for prediction.

All in all we predicted 12 million new matches between 1900 and 1910. We took a random sample of 30 from 3 different states and found them to be 90% accurate.
